```python
# Variables Task 1
k = 93 # ùúâ2
p = 0.49 # ùúâ3
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import nbinom

def plot_negative_binomial(k, p):
    x = np.arange(nbinom.ppf(0.01, k, p), nbinom.ppf(0.99, k, p))
    y = nbinom.pmf(x, k, p)

    plt.plot(x, y, 'bo', ms=8)
    plt.vlines(x, 0, y, colors='b', lw=5)
    plt.title('Negative Binomial Distribution (k={}, p={})'.format(k, p))
    plt.xlabel('Number of Failures')
    plt.ylabel('Probability')
    plt.show()
```

```python
# Execute plot function

plot_negative_binomial(k, p)
```
